**Parte 2:Carregando e Pré-processando os Dados**

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

Carregar o conjunto de dados [SMS Spam Collection](https://archive.ics.uci.edu/dataset/228/sms+spam+collection)

```python
df = pd.read_csv('SMSSpamCollection', sep='\t', header=None, names=['label', 'message'])
```

Exibir as primeiras linhas do DataFrame para entender os dados

```python
print(df.head())
```

Converter mensagens de texto em vetores de contagem

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['message'])
y = df['label']
```

Dividir os dados em conjuntos de treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Parte 3:Treinamento do Modelo**

```python
#Treinar o modelo de Naive Bayes
model = MultinomialNB()
model.fit(X_train, y_train)
```

**Parte 4:Avaliação do Modelo**

```python
#Fazer previsões
y_pred = model.predict(X_test)

#Avaliar o modelo
print("Accuracy:", accuracy_score(y_test, y_pred))
```

**Parte 5:Testar com Novas Mensagens**

```python
#Testar com novas mensagens
new_messages = ['Congratulations! You won a car. Click here for more details.', 'Can we have lunch tomorrow?', 'Special offer for you!', 'You password has expired. Click here to reset your password.']
new_X = vectorizer.transform(new_messages)
predictions = model.predict(new_X)

for message, label in zip(new_messages, predictions):
    print(f'Mensagem: "{message}" é classificada como {label}')
```
